# Create Movie Table

```python
df = pd.read_csv("2022_movies_final_merged.csv")
df.columns
```

```python
df = df.rename(columns={"id":"movieID","movie": "name", "storyPlot": "description",
                       "release_date":"Release_Date",
                       "small_posters":"Image_URL","genres":"GenreFull",
                       })
```

```python
# Randomly create 0 and 1
import random
import math
my_list = ["N"]*math.floor(len(df)/2) + ["Y"]*(len(df)-math.floor(len(df)/2))
random.shuffle(my_list)
```

```python
# add random shuffled 0 and 1 to new column "Active"
df["Active"] = my_list
```

```python
df
```

```python
# just leave the genre column with the first genre of the movie for easier reference in MySQL
genre_col = []
for i, item in df.iterrows():
    genre_col.append(item["GenreFull"].split(",")[0])
df["genre"] = genre_col
```

# Create Genre table

```python

genre = []
for item in df["GenreFull"]:
    if len(item) > 1:
        l = [x.strip() for x in item.split(',')]
    else:
        l = item
    genre.append(l)
```

```python
import itertools
merged = sorted(list(set(list(itertools.chain.from_iterable(genre)))))
merged
```

```python
# open mapping file
import json
with open('genre_descriptions.json') as user_file:
    file_contents = user_file.read()

parsed_json = json.loads(file_contents)
parsed_json['Action']
```

```python
df_genre = pd.DataFrame(merged, columns=["name"]) 
df_genre
```

```python
# find the corresponding description
description = []
for index,item in df_genre.iterrows():
    description.append(parsed_json[item["name"]])
df_genre["description"] = description
df_genre
```

```python
df_genre.to_csv("genre_MySQL.csv", index=False)
```

# Update "movie_MySQL.csv" with the new genreId based on df_genre

```python
df["genre"] = df["genre"].str.strip()
df["genre"]
```

```python

# create GenreId on the movie_MySQL.csv
GenreId = []
for i,item in df.iterrows():
#     df_genre.iloc[select_indices].index.tolist()
    idx = df_genre.index[df_genre['name']==item["genre"]] + 1
    print(idx.tolist())
    # because GenreId contains list of list, just take the element in the inner list
    GenreId.append(idx[0])
df["GenreId"] = GenreId
```

```python
# Release_date modify to the format like in the assignment: "yyyy/mm/dd" 
monthMap = {"Jan":"01", "Feb":"02", "Mar":"03", "Apr":"04", "May":"05", "Jun":"06", "Jul":"07",
           "Aug":"08", "Sep":"09","Oct":"10", "Nov":"11", "Dec":"12"}
full_date_list = []
for i,item in df.iterrows():
    year= item["Release_Date"].split(", ")[-1]
    day= item["Release_Date"].split(", ")[0].split(" ")[1] 
    # if the day is a single digit, append 0 infront to make it double digit
    if len(day) < 2:
        day = "0" + day
    month_word = item["Release_Date"].split(' ')[0]
    month= monthMap[month_word]
    # join together to form the new string we want "yyyy/mm/dd"
    joined = year + "/" + month + "/" + day
    full_date_list.append(joined)
df["Release_Date"] = full_date_list
```

```python
# Create the DateInserted column
import time
from datetime import datetime, timedelta

# make a list of DateInserted the same length as the dataframe
# current timestamp
x = time.time()# convert to datetime
dt = datetime.fromtimestamp(x)
# the list to add to df
overallDateInserted = []
print(dt)
# the SQL datetime format
for i in range(len(df)):
    delta = 24 + i
    result = dt + timedelta(seconds=delta)
    DateInserted = str(result).split(".")[0]
    overallDateInserted.append(DateInserted)
df["DateInserted"] = overallDateInserted
```
